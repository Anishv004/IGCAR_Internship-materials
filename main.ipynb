```python
import re
df['Split_Names'] = df['BrowserDetails'].apply(lambda x: re.split(r'\s+(?![^()]*\))', x))
```

```python
def split_items(item_list):
    new_items = []
    for item in item_list:
        if '(' in item:
            items_inside_parentheses = item.split('(')[1].replace(')', '').split(';')
            new_items.extend([i.strip() for i in items_inside_parentheses])
        else:
            new_items.extend([i.strip() for i in item.split(';')])
    return new_items
df['Split_Names'] = df['Split_Names'].apply(split_items)
```

```python
def split_items_by_comma(item_list):
    new_items = [item.split(',') for item in item_list]
    return [i.strip() for sublist in new_items for i in sublist]

df['Split_Names'] = df['Split_Names'].apply(split_items_by_comma)
```

```python
df['Split_Names'][0]
```

```python
unique_values_Browser = set(item for sublist in df['Split_Names'] for item in sublist)
for value in unique_values_Browser:
    df[value] = df['Split_Names'].apply(lambda x: value in x)
```
